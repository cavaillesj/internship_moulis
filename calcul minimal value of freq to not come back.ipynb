```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fact(n):
    f=1
    for k in range(1, n):
        f*=k
    return f

combin = lambda n,k: fact(n)//(fact(k)*fact(n-k))
```

```python
def proba_sup_2(p, n):
    s = 0
    for k in range(n):
        s += (n-k-1)*combin(n,k)*(-p)**(n-k)
    return s
```

### depend on the choice of n ...

```python
n = 10

gamma = 0.95
P = [0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8, 2]
Proba = np.zeros_like(P)
for i, p in enumerate(P):
    Proba[i] = proba_sup_2(p, n)
plt.semilogy(P, Proba, "*-", label = "proba")
plt.semilogy(P, [gamma]*len(P), label = "gamma")
plt.legend()
plt.show()
```
